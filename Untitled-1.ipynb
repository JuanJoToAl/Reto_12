```python
# Se declaran variables
contenido : str
texto_vocales_minusculas : str
vocales : str
cantidad_a : int
cantidad_e : int
cantidad_i : int
cantidad_o : int
cantidad_u : int
suma : int
variables : list
# Se abre archivo 
archivo = open("archivo.txt")

def solo_vocales(archivo):
    # Se almacena contenido del archivo en formato str
    contenido = archivo.read()
    # Se crea variable para almacenar las vocales
    texto_vocales_minusculas = ""
    # Se crea variable para filtrar contenido de variable "contenido" 
    vocales = "aeiouAEIOU"
    
    for caracter in contenido:
        # Se comprueba si caracter es una vocal
        if caracter in vocales:
            # Si se cumple condución el caracter se añade a variable
            texto_vocales_minusculas += caracter.lower()
    # Se retorna str con todas las vocales del archivo
    return texto_vocales_minusculas

def contar_vocales(texto_vocales_minusculas):

    cantidad_a = 0
    cantidad_e = 0
    cantidad_i = 0
    cantidad_o = 0
    cantidad_u = 0

    for caracter in texto_vocales_minusculas:
        # Se suman indidualmente las vocales del texto
        cantidad_a += caracter.count("a")
        cantidad_e += caracter.count("e")
        cantidad_i += caracter.count("i")
        cantidad_o += caracter.count("o")
        cantidad_u += caracter.count("u")
    # Se retorna cantidad total de cada vocal en el archivo
    return cantidad_a, cantidad_e, cantidad_i, cantidad_o, cantidad_u

if __name__ == "__main__":
    # Se llama función y se retorna texto con solo las vocales 
    texto_vocales_minusculas = solo_vocales(archivo)
    # Se llama función y se retorna la cantidad de cada vocal en el texto
    cantidad_a, cantidad_e, cantidad_i, cantidad_o, cantidad_u = contar_vocales(
                                                                                texto_vocales_minusculas)
    variables = [cantidad_a, cantidad_e, cantidad_i, cantidad_o, cantidad_u]
    # Se obtiene cantidad total de vocales en el texto
    suma = sum(variables)
    # Se imprimen resultados
    print(f"La vocal a aparece {cantidad_a} veces")
    print(f"La vocal e aparece {cantidad_e} veces")
    print(f"La vocal i aparece {cantidad_i} veces")
    print(f"La vocal o aparece {cantidad_o} veces")
    print(f"La vocal u aparece {cantidad_u} veces")
    print(f"La cantidad total de vocales {suma}")
# Se cierra archivo
archivo.close()
```

```python
# Se declaran variables
contenido : str
texto_consonantes_minusculas : str
vocales : str
suma : int
total_consonantes : int
variables : list
# Se abre archivo 
archivo = open("archivo.txt")

def solo_consonantes(archivo):
    # Se almacena contenido del archivo en formato str
    contenido = archivo.read()
    # Se crea variable para almacenar las consonantes
    texto_consonantes_minusculas = ""
    # Se crea variable para filtrar contenido de variable "contenido" 
    vocales = "aeiouAEIOU"
    for caracter in contenido:
        # Se convierten los caracteres del archivo a minuscula
        caracter = caracter.lower()
        # Comprobación si caracter es una letra minúscula con ASCII
        if 98 <= ord(caracter) and ord(caracter) <= 122:
            # Se comprueba si el caracter es una consonante
            if caracter not in  vocales:
                # Si se cumple condución el caracter se añade a variable
                texto_consonantes_minusculas += caracter
    # Se retorna str con todas las consonantes del archivo
    return texto_consonantes_minusculas

def contar_consonantes(texto_consonantes_minusculas):
    # Se cuenta la cantidad todal de consonantes en el archivo
    total_consonantes = len(texto_consonantes_minusculas)
    # Se retorna la cantidad de consonantes en el archivo
    return total_consonantes

if __name__ == "__main__":
    # Se llama función y se retorna texto con solo las consonantes 
    texto_consonantes_minusculas = solo_consonantes(archivo)
    # Se llama función y se retorna la cantidad de consonantes en el texto
    total_consonantes = contar_consonantes(texto_consonantes_minusculas)
    # Se imprime resultado con la cantidad todal de consonantes 
    print(f"La cantidad total de consonantes es {total_consonantes}")
# Se cierra archivo
archivo.close()
```

```python
def palabras_archivo(contenido):
    texto_para_lista = ""
    lista = []
    for caracter in contenido:
        caracter.lower()
        if 97 <= ord(caracter) and ord(caracter) <= 122:
            texto_para_lista += caracter

        elif len(texto_para_lista) == 1:
            texto_para_lista = ""

        else:
            if texto_para_lista:  # Verifique si texto_nuevo no está vacío antes de agregar
                lista.append(texto_para_lista)
                texto_para_lista = ""  # Limpie texto_nuevo después de agregar a la lista

    if texto_para_lista:  # Verifique si texto_nuevo no está vacío antes de agregar
        lista.append(texto_para_lista)
        texto_para_lista = ""  # Limpie texto_nuevo después de agregar a la lista

    return lista

def contar_palabras(lista):
    palabras = []
    numeros = []

    for palabra in lista:
        if palabra.capitalize() not in palabras:
            palabras.append(palabra.capitalize())
            numeros.append(lista.count(palabra))
    return palabras, numeros

def ordenar_palabras(palabras, numeros):
    veces_palabras = []
    contador = 0

    while contador < 50:

        indice = numeros.index(max(numeros))

        veces_palabras.append(palabras[indice])
        veces_palabras.append(max(numeros))

        numeros.remove(max(numeros))
        numeros.insert(indice, 0)

        contador += 1
    
        

    return veces_palabras

def ordenar_palabras_texto_corto(palabras, numeros):
    veces_palabras = []
    contador = 0

    while contador < len(numeros):

        indice = numeros.index(max(numeros))

        veces_palabras.append(palabras[indice])
        veces_palabras.append(max(numeros))

        numeros.remove(max(numeros))
        numeros.insert(indice, 0)

        contador += 1
    
    return veces_palabras

if __name__ == "__main__":
    archivo = open("archivo.txt")
    contenido = archivo.read()
    # Se llama función y se retorna texto con solo las consonantes 
    lista = palabras_archivo(contenido)
    # Se llama función y se retorna la cantidad de consonantes en el texto

    palabras, numeros = contar_palabras(lista)

    if 50 <= len(numeros):
        veces_palabras = ordenar_palabras(palabras, numeros)
    else:
        veces_palabras = ordenar_palabras_texto_corto(palabras, numeros)
    print("Listado de las 50 palabras que más se repiten")
    print("----------------------------------------------")
    for x in range(0, len(veces_palabras), 2):
        print(f"Palabra: {veces_palabras[x]} | Apariciones: {veces_palabras[x + 1]}")
        print("---------------------------------------------")

    archivo.close()
```

```python
import re

def contar_palabras(archivo):
  """
  Cuenta la frecuencia de las palabras en un archivo de texto.

  Argumentos:
    archivo (str): La ruta del archivo de texto.

  Devuelve:
    dict: Un diccionario donde las claves son las palabras y los valores son la cantidad de veces que aparecen.
  """
  with open(archivo, 'r') as f:
    contenido = f.read()

  # Eliminar signos de puntuación, mayúsculas y convertir a minúsculas
  texto_limpio = re.sub(r'[^\w\s]', '', contenido).lower()
  palabras = texto_limpio.split()

  # Contar la frecuencia de las palabras
  frecuencias = {}
  for palabra in palabras:
    if palabra in frecuencias:
      frecuencias[palabra] += 1
    else:
      frecuencias[palabra] = 1

  return frecuencias

def obtener_50_mas_repetidas(frecuencias):
  """
  Obtiene las 50 palabras más repetidas de un diccionario de frecuencias.

  Argumentos:
    frecuencias (dict): Un diccionario donde las claves son las palabras y los valores son la cantidad de veces que aparecen.

  Devuelve:
    list: Una lista con las 50 palabras más repetidas en orden descendente de frecuencia.
  """
  # Ordenar las palabras por frecuencia en orden descendente
  palabras_ordenadas = sorted(frecuencias.items(), key=lambda x: x[1], reverse=True)

  # Obtener las 50 palabras más repetidas
  mas_repetidas = []
  for palabra, frecuencia in palabras_ordenadas[:50]:
    mas_repetidas.append((palabra, frecuencia))

  return mas_repetidas

if __name__ == "__main__":

  archivo = "archivo.txt" # Reemplazar con la ruta del archivo de texto

  # Contar las frecuencias de las palabras
  frecuencias = contar_palabras(archivo)

  # Obtener las 50 palabras más repetidas
  mas_repetidas = obtener_50_mas_repetidas(frecuencias)

  # Imprimir las 50 palabras más repetidas
  for palabra, frecuencia in mas_repetidas:
    print(f"{palabra}: {frecuencia}")

```
